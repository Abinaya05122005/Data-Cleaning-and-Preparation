```python
import pandas as pd
```

```python
df = pd.read_excel("/content/Dataset for Data Analytics.xlsx")
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
df["CouponCode"] = df["CouponCode"].fillna("No Coupon")

print(df["CouponCode"].isnull().sum())
```

```python
duplicates = df.duplicated().sum()

print("Duplicate Rows:", duplicates)
```

```python
df = df.drop_duplicates()

print("Total Records:", len(df))
```

```python
duplicate_ids = df["OrderID"].duplicated().sum()

print("Duplicate Order IDs:", duplicate_ids)
```

```python
df["Date"] = pd.to_datetime(df["Date"])

print(df["Date"].dtype)
```

```python
print(df.dtypes)
```

```python
df.to_excel("cleaned_Dataset_for_Data_Analytics.xlsx", index=False)

print("Dataset cleaned successfully!")
display(df)
```
